```python
import os
from PIL import Image
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import plot_tree
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
)
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import time

folders = [
    'DM_dataset_hw1/battery',
    'DM_dataset_hw1/dc_volt_src',
    'DM_dataset_hw1/diode',
    'DM_dataset_hw1/gnd',
    'DM_dataset_hw1/inductor',
    'DM_dataset_hw1/resistor',
    'DM_dataset_hw1/voltmeter'
]

images = []

for folder_path in folders:
    for folder in os.listdir(folder_path):
        image_path = os.path.join(folder_path, folder)
        image = Image.open(image_path)
        image_np = np.asarray(image)

        image_28x28 = np.array(Image.fromarray(image_np).resize((28, 28)))

        image_28x28 = image_28x28 / 255.0

        image_1d = image_28x28.reshape(-1)

        images.append(image_1d)

image_shapes = [image.shape for image in images]

min_shape = min(image_shapes, key=lambda x: x[0])

images_resized = [image[:min_shape[0]] for image in images]

record_matrix = np.vstack(images_resized)

labels = [0] * len(os.listdir('DM_dataset_hw1/battery')) + \
         [1] * len(os.listdir('DM_dataset_hw1/dc_volt_src')) + \
         [2] * len(os.listdir('DM_dataset_hw1/diode')) + \
         [3] * len(os.listdir('DM_dataset_hw1/gnd')) + \
         [4] * len(os.listdir('DM_dataset_hw1/inductor')) + \
         [5] * len(os.listdir('DM_dataset_hw1/resistor')) + \
         [6] * len(os.listdir('DM_dataset_hw1/voltmeter'))

X_train, X_test, y_train, y_test = train_test_split(record_matrix, labels, test_size=0.2, random_state=42)

k_values = [3, 5, 7]
k_fold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for train_indices, test_indices in k_fold.split(record_matrix, labels):
    X_train, X_test = record_matrix[train_indices], record_matrix[test_indices]
    y_train, y_test = np.array(labels)[train_indices], np.array(labels)[test_indices]

    for k in k_values:
        knn_euclidean = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
        start_time = time.time()
        knn_euclidean.fit(X_train, y_train)
        training_time_knn_euclidean = time.time() - start_time

        start_time = time.time()
        y_pred_euclidean = knn_euclidean.predict(X_test)
        testing_time_knn_euclidean = time.time() - start_time

        accuracy_knn_euclidean = accuracy_score(y_test, y_pred_euclidean)
        precision_knn_euclidean = precision_score(y_test, y_pred_euclidean, average='weighted')
        recall_knn_euclidean = recall_score(y_test, y_pred_euclidean, average='weighted')
        f1_knn_euclidean = f1_score(y_test, y_pred_euclidean, average='weighted')

        print(f"KNN (Euclidean) - k={k} Metrics:")
        print(f"Accuracy: {accuracy_knn_euclidean:.4f}")
        print(f"Precision: {precision_knn_euclidean:.4f}")
        print(f"Recall: {recall_knn_euclidean:.4f}")
        print(f"F-measure: {f1_knn_euclidean:.4f}")
        print(f"Training Time: {training_time_knn_euclidean:.4f} seconds")
        print(f"Testing Time: {testing_time_knn_euclidean:.4f} seconds")

        conf_matrix_knn_euclidean = confusion_matrix(y_test, y_pred_euclidean)
        disp_knn_euclidean = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_knn_euclidean, display_labels=np.unique(labels))
        disp_knn_euclidean.plot(cmap=plt.cm.Blues)
        plt.title(f'Confusion Matrix - KNN (Euclidean) - k={k}')
        plt.show()

        print("**************************************************************")

for k in k_values:
    knn_manhattan = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    
    for train_indices, test_indices in k_fold.split(record_matrix, labels):
        X_train, X_test = record_matrix[train_indices], record_matrix[test_indices]
        y_train, y_test = np.array(labels)[train_indices], np.array(labels)[test_indices]

        start_time = time.time()
        knn_manhattan.fit(X_train, y_train)
        training_time_knn_manhattan = time.time() - start_time

        start_time = time.time()
        y_pred_manhattan = knn_manhattan.predict(X_test)
        testing_time_knn_manhattan = time.time() - start_time

        accuracy_knn_manhattan = accuracy_score(y_test, y_pred_manhattan)
        precision_knn_manhattan = precision_score(y_test, y_pred_manhattan, average='weighted')
        recall_knn_manhattan = recall_score(y_test, y_pred_manhattan, average='weighted')
        f1_knn_manhattan = f1_score(y_test, y_pred_manhattan, average='weighted')

        print(f"KNN (Manhattan) - k={k} Metrics:")
        print(f"Accuracy: {accuracy_knn_manhattan:.4f}")
        print(f"Precision: {precision_knn_manhattan:.4f}")
        print(f"Recall: {recall_knn_manhattan:.4f}")
        print(f"F-measure: {f1_knn_manhattan:.4f}")
        print(f"Training Time: {training_time_knn_manhattan:.4f} seconds")
        print(f"Testing Time: {testing_time_knn_manhattan:.4f} seconds")

        conf_matrix_knn_manhattan = confusion_matrix(y_test, y_pred_manhattan)
        disp_knn_manhattan = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_knn_manhattan, display_labels=np.unique(labels))
        disp_knn_manhattan.plot(cmap=plt.cm.Blues)
        plt.title(f'Confusion Matrix - KNN (Manhattan) - k={k}')
        plt.show()

        print("**************************************************************")

decision_tree = DecisionTreeClassifier()
start_time = time.time()
decision_tree.fit(X_train, y_train)
training_time_decision_tree = time.time() - start_time

start_time = time.time()
y_pred_decision_tree = decision_tree.predict(X_test)
testing_time_decision_tree = time.time() - start_time

accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
precision_decision_tree = precision_score(y_test, y_pred_decision_tree, average='weighted')
recall_decision_tree = recall_score(y_test, y_pred_decision_tree, average='weighted')
f1_decision_tree = f1_score(y_test, y_pred_decision_tree, average='weighted')

print("Decision Tree Metrics:")
print(f"Accuracy: {accuracy_decision_tree:.4f}")
print(f"Precision: {precision_decision_tree:.4f}")
print(f"Recall: {recall_decision_tree:.4f}")
print(f"F-measure: {f1_decision_tree:.4f}")
print(f"Training Time: {training_time_decision_tree:.4f} seconds")
print(f"Testing Time: {testing_time_decision_tree:.4f} seconds")

conf_matrix_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)
disp_decision_tree = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_decision_tree, display_labels=np.unique(labels))
disp_decision_tree.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix - Decision Tree')
plt.show()
    
plt.figure(figsize=(20, 10))
plot_tree(decision_tree, filled=True, feature_names=[str(i) for i in range(record_matrix.shape[1])], class_names=[str(i) for i in np.unique(labels)])
plt.title('Decision Tree')
plt.show()

print("**************************************************************")

data = np.hstack((record_matrix, np.array(labels).reshape(-1, 1)))

X = data[:, :-1]
y = data[:, -1]

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    start_time = time.time()
    decision_tree_entropy = DecisionTreeClassifier(criterion='entropy')
    decision_tree_entropy.fit(X_train, y_train)
    end_time = time.time()

    y_pred_decision_tree_entropy = decision_tree_entropy.predict(X_test)

    accuracy_entropy = accuracy_score(y_test, y_pred_decision_tree_entropy)
    precision_entropy = precision_score(y_test, y_pred_decision_tree_entropy, average='weighted')
    recall_entropy = recall_score(y_test, y_pred_decision_tree_entropy, average='weighted')
    f1_entropy = f1_score(y_test, y_pred_decision_tree_entropy, average='weighted')
    training_time_entropy = end_time - start_time

    start_time = time.time()
    y_pred_test_decision_tree_entropy = decision_tree_entropy.predict(X_test)
    testing_time_entropy = time.time() - start_time

    print("Decision Tree with Entropy:")
    print(f"Accuracy: {accuracy_entropy:.4f}")
    print(f"Precision: {precision_entropy:.4f}")
    print(f"Recall: {recall_entropy:.4f}")
    print(f"F1 Score: {f1_entropy:.4f}")
    print(f"Training Time: {training_time_entropy:.4f} seconds")
    print(f"Testing Time: {testing_time_entropy:.4f} seconds")

    conf_matrix_decision_tree_entropy = confusion_matrix(y_test, y_pred_decision_tree_entropy)
    disp_decision_tree_entropy = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_decision_tree_entropy, display_labels=np.unique(labels))
    disp_decision_tree_entropy.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix - Decision Tree (Entropy)')
    plt.show()

    print("**************************************************************")

for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    start_time = time.time()
    decision_tree_gini = DecisionTreeClassifier(criterion='gini')
    decision_tree_gini.fit(X_train, y_train)
    end_time = time.time()

    y_pred_decision_tree_gini = decision_tree_gini.predict(X_test)

    accuracy_gini = accuracy_score(y_test, y_pred_decision_tree_gini)
    precision_gini = precision_score(y_test, y_pred_decision_tree_gini, average='weighted')
    recall_gini = recall_score(y_test, y_pred_decision_tree_gini, average='weighted')
    f1_gini = f1_score(y_test, y_pred_decision_tree_gini, average='weighted')
    training_time_gini = end_time - start_time

    start_time = time.time()
    y_pred_test_decision_tree_gini = decision_tree_gini.predict(X_test)
    testing_time_gini = time.time() - start_time

    print("\nDecision Tree with Gini Index:")
    print(f"Accuracy: {accuracy_gini:.4f}")
    print(f"Precision: {precision_gini:.4f}")
    print(f"Recall: {recall_gini:.4f}")
    print(f"F1 Score: {f1_gini:.4f}")
    print(f"Training Time: {training_time_gini:.4f} seconds")
    print(f"Testing Time: {testing_time_gini:.4f} seconds")

    conf_matrix_decision_tree_gini = confusion_matrix(y_test, y_pred_decision_tree_gini)
    disp_decision_tree_gini = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_decision_tree_gini, display_labels=np.unique(labels))
    disp_decision_tree_gini.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix - Decision Tree (Gini Index)')
    plt.show()

    print("**************************************************************")

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    
for train_index, test_index in skf.split(record_matrix, labels):
    X_train, X_test = record_matrix[train_index], record_matrix[test_index]
    y_train, y_test = np.array(labels)[train_index], np.array(labels)[test_index]

    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)
    y_pred_decision_tree = decision_tree.predict(X_test)

    naive_bayes = GaussianNB()

    start_time = time.time()
    naive_bayes.fit(X_train, y_train)
    training_time = time.time() - start_time

    start_time = time.time()
    y_pred_naive_bayes = naive_bayes.predict(X_test)
    testing_time = time.time() - start_time

    accuracy_naive_bayes = accuracy_score(y_test, y_pred_naive_bayes)
    precision_naive_bayes = precision_score(y_test, y_pred_naive_bayes, average='weighted')
    recall_naive_bayes = recall_score(y_test, y_pred_naive_bayes, average='weighted')
    f1_naive_bayes = f1_score(y_test, y_pred_naive_bayes, average='weighted')

    print(f"Accuracy (Naive Bayes): {accuracy_naive_bayes:.4f}")
    print(f"Precision (Naive Bayes): {precision_naive_bayes:.4f}")
    print(f"Recall (Naive Bayes): {recall_naive_bayes:.4f}")
    print(f"F1 Score (Naive Bayes): {f1_naive_bayes:.4f}")
    print(f"Training Time (Naive Bayes): {training_time:.4f} seconds")
    print(f"Testing Time (Naive Bayes): {testing_time:.4f} seconds")

    conf_matrix_naive_bayes = confusion_matrix(y_test, y_pred_naive_bayes)
    disp_naive_bayes = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_naive_bayes, display_labels=np.unique(labels))
    disp_naive_bayes.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix - Naive Bayes')
    plt.show()

print("**************************************************************")

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

data = np.hstack((record_matrix, np.array(labels).reshape(-1, 1)))

X = data[:, :-1]
y = data[:, -1]
    
for train_index, test_index in kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    svm_classifier = SVC()
    start_time_svm_train = time.time()
    svm_classifier.fit(X_train, y_train)
    end_time_svm_train = time.time()
    start_time_svm_test = time.time()
    y_pred_svm = svm_classifier.predict(X_test)
    end_time_svm_test = time.time()

    accuracy_svm = accuracy_score(y_test, y_pred_svm)
    precision_svm = precision_score(y_test, y_pred_svm, average='weighted')
    recall_svm = recall_score(y_test, y_pred_svm, average='weighted')
    f1_svm = f1_score(y_test, y_pred_svm, average='weighted')

    conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
    disp_svm = ConfusionMatrixDisplay(confusion_matrix=conf_matrix_svm, display_labels=np.unique(labels))
    disp_svm.plot(cmap=plt.cm.Blues)
    plt.title('Confusion Matrix - SVM')
    plt.show()

    print(f"Accuracy: {accuracy_svm:.4f}")
    print(f"Precision: {precision_svm:.4f}")
    print(f"Recall: {recall_svm:.4f}")
    print(f"F1 Score: {f1_svm:.4f}")
    print(f"Training Time: {end_time_svm_train - start_time_svm_train:.4f} seconds")
    print(f"Testing Time: {end_time_svm_test - start_time_svm_test:.4f} seconds")
    print("**************************************************************")
```
